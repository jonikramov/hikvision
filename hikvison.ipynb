```python
def abc_analysis(total):
    if total < 80:
        return 'A'
    elif total < 95:
        return 'B'
    else:
        return 'C'    
```

# Год 2023

```python
df_23 = df.query('date < "2024-01-01"')
```

```python
df_23.id_product.nunique()
```

```python
df_23.product_name.nunique()
```

```python
df_23.head()
```

```python
df_23.date.describe()
```

```python
total_revenue_23 = df_23.revenue.sum()
total_revenue_23
```

### Группируем по id_product и product_name

#### Датасэт df_23_product_revenue_a

```python
df_23_product_revenue_a = df_23.groupby(['id_product', 'product_name'], as_index=0).agg({'revenue':'sum'})
df_23_product_revenue_a.head()
```

```python
df_23_product_revenue_a = df_23_product_revenue_a.sort_values('revenue', ascending=False)
```

```python
df_23_product_revenue_a['usd'] = round(df_23_product_revenue_a['revenue'] / 12500, 2)
```

```python
df_23_product_revenue_a['percent_total_revenue'] = round(df_23_product_revenue_a['revenue'] / df_23_product_revenue_a.revenue.sum() * 100, 2)
```

```python
df_23_product_revenue_a['expanding_total_revenue'] = df_23_product_revenue_a['percent_total_revenue'].expanding().sum()
```

```python
df_23_product_revenue_a['group'] = df_23_product_revenue_a['expanding_total_revenue'].apply(abc_analysis)
```

```python
df_23_product_revenue_a
```

```python
df_23_product_revenue_a.groupby('group', as_index=0).agg({'id_product': 'nunique', 'product_name': 'nunique', 'revenue': 'sum'})
```

### Группируем по id_product 2023 год

#### Датасэт df_23_product_revenue_b

```python
df_23_product_revenue_b = df_23.groupby(['id_product'], as_index=0).agg({'revenue':'sum'})
df_23_product_revenue_b.head()
```

```python
df_23_product_revenue_b = df_23_product_revenue_b.sort_values('revenue', ascending=False)
```

```python
df_23_product_revenue_b['usd'] = round(df_23_product_revenue_b['revenue'] / 12500, 2)
```

```python
df_23_product_revenue_b['percent_total_revenue'] = round(df_23_product_revenue_b['revenue'] / df_23_product_revenue_b.revenue.sum() * 100, 2)
```

```python
df_23_product_revenue_b['expanding_total_revenue'] = df_23_product_revenue_b['percent_total_revenue'].expanding().sum()
```

```python
df_23_product_revenue_b['group'] = df_23_product_revenue_b['expanding_total_revenue'].apply(abc_analysis)
```

```python
df_23_product_revenue_b
```

```python
df_23_product_revenue_b.groupby('group', as_index=0).agg({'id_product': 'count', 'revenue': 'sum'})
```
